```python
# -*- coding: utf-8 -*-
"""
1_zacks_exprep_hist.ipynb

Assuming all the Exp_Rep files are downloaded (.ex5 files), 
gather all ex.5 files into history 

@author: clj585
"""
import pandas as pd
import ftplib
import os, glob
import datatable as dt
from datetime import datetime, timedelta
import zipfile
from datatable import Frame
import pyarrow.parquet as pq

"""
Collect all files

"""
# ToDo: change pathname to OneDrive\Zacks\data\ex5files
dir_pth = "C:\\Users\clj585\Dropbox\Zacks\complete_raw_data\ex5files"   

# Extract from Dropbox all the *.ex5 files 
os.chdir(f"{dir_pth}")
files = []                              # save each datatable data frame
for filename in glob.glob("*.ex5"):
    print(os.path.join(dir_pth, filename))
    rawfile = dt.fread(os.path.join(dir_pth, filename),
                      header=False)
    rawfile['asofdate'] = str(filename)[0:-4]
    files.append(rawfile)

exp_report = dt.rbind(files, force=True)

"""
Save most current history as parquet

"""
curr_day = datetime.now().day
curr_month = datetime.now().month
curr_yr = datetime.now().year

save_pth = "C:\\Users\clj585\Dropbox\Zacks"            # ToDo: change pathname to OneDrive\Zacks\data\

new_arrow_table = exp_report.to_arrow()
pq.write_table(new_arrow_table, f"{save_pth}\exprep_hist_asof_{curr_month}_{curr_day}_{curr_yr}.parquet")

print("Ready to go")
```
