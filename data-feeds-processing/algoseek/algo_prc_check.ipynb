```python
def my_ret(df):
    df_ret = df
    df_ret['lag_cpf'] = df_ret['CumulativePriceFactor'].shift(1)
    df_ret['lag_af'] = df_ret['AdjustmentFactor'].shift(1)
    df_ret['lead_cpf'] = df_ret['CumulativePriceFactor'].shift(-1)
    df_ret['lead_af'] = df_ret['AdjustmentFactor'].shift(-1)
    df_ret['adj_closeprc'] = df_ret['CumulativePriceFactor'] * df_ret['ClosePrice']
    df_ret['adj_closeprc'] = np.where((df_ret['AdjustmentFactor'].isna()==True) | 
                                       (df_ret['AdjustmentFactor'] == 1/df_ret['CumulativePriceFactor']), 
                                      df_ret['adj_closeprc'], 
                                      df_ret['adj_closeprc']*df_ret['AdjustmentFactor']
                                     )
    df_ret['lag_adj_closeprc'] = df_ret['adj_closeprc'].shift(1)
    df_ret['lag_adj_closeprc'] = np.where(df_ret['CumulativePriceFactor'] != df_ret['lag_cpf'], 
                                        df_ret['lag_adj_closeprc'] * df_ret['lag_cpf'], 
                                          df_ret['lag_adj_closeprc']
    )
    
    df_ret['my_ret'] = df_ret['adj_closeprc']/df_ret['lag_adj_closeprc'] - 1
    return df_ret
    
banned_altria_ret = my_ret(banned)
banned_altria_ret
```
