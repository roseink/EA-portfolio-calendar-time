```python
"""
File: algoseek_parquet_processing.ipynb

Simple conversion pre-processing columns first from CSV, 
then save each daily ochl file from s@ bucket as parquet file. 

Data outputs located in temporary diretory in algoseek/Data 
folder. 

Last updated: 5/18/22

"""
# Imports here 
import os
import glob
import numpy as np
import pandas as pd
from datetime import datetime

# Location of pulled CSVs from S3 Bucket
csv_dir = "C://Users/clj585/OneDrive - Northwestern University/data_feeds/algoseek/Data/"

# Go to directory and loop over all csvs
os.chdir(csv_dir)
for filen in glob.glob("*.csv"):
    # Each csv file starts with date, begins with year 2000s (data starts after 2007)
    if filen.startswith("2"):
        print(filen)
        df = pd.read_csv(os.path.join(filen))
        
        # Change and fix column types 
        df[['TradeDate', 'Ticker', 'Name', 'PrimaryExchange', 'ISIN',
            'AdjustmentReason']] = df[['TradeDate', 'Ticker', 'Name', 'PrimaryExchange', 'ISIN',
                                       'AdjustmentReason']].astype("string")
        # Adjust dates 
        df['TradeDate'] = pd.to_datetime(df['TradeDate'])
        
        # Save and form file name
        numstr = str(filen)[:-4]
        print(f"{csv_dir}temp_parquet_daily/{numstr}.parquet")
        df.to_parquet(f"{csv_dir}temp_parquet_daily/{numstr}.parquet")
        
```

```python
"""
From pre-processed parquet files, further process columns 
and rename according to our convention in data_feeds. 

"""
# Change directory to location of pre-processed temp parquet files
parq_dir = csv_dir+"/temp_parquet_daily/"
os.chdir(parq_dir)

# Loop over each individual raw file name
for rawfile in glob.glob("*.parquet"):
    blah = pd.read_parquet(os.path.join(rawfile))
    print(rawfile)
    
    # Change and define column names by the following mapping 
    colmap = {'TradeDate' : "date", 'SecId':"algo_id", 'Ticker':'hticker', "Name":"name", 
             "ISIN":"isin"}
    blah = blah.rename(columns=colmap)
    
    # Cusips are historical, point in time, get from ISIN
    blah['hcusip'] = blah['isin'].str[2:10]
    blah['hcusip9'] = blah['isin'].str[2:11]

    # Process the ticker column, share classes are denoted with "."
    blah[['root', 'suffix']] = blah['hticker'].str.split('[ /.]', n=1, expand=True)
    blah[['root', 'suffix']] = blah[['root', 'suffix']].fillna('')

    # Drop auxiliary columns
    print(blah.columns)
    
    # Save finished and renamed data frame as parquet 
    numstr = str(rawfile)[:-8]
    print(csv_dir+"renamed_parquet_daily/"+numstr+"_renamed.parquet")
    blah.to_parquet(csv_dir+"renamed_parquet_daily/"+numstr+"_renamed.parquet")
```
