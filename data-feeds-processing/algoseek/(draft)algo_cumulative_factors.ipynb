```python
# Imported libraries 
import boto3
import pandas as pd 
import numpy as np
from datetime import datetime, timedelta, date
from dateutil.relativedelta import relativedelta
import time
import exchange_calendars as xcals
import sqlite3
import glob, os

# Have to join all daily factors files together 
onedrive_pth = "C://Users/clj585/OneDrive - Northwestern University/"
factors_pth = "data_feeds/algoseek/Data/temp_daily_adj_factors/"

os.chdir(onedrive_pth+factors_pth)

list_daily_files = []

csv_files = glob.glob("*.csv")
for fname in csv_files:
    print(fname)
    temp_df = pd.read_csv(os.path.join(fname))
    list_daily_files.append(temp_df)
    
fac_df = pd.concat(list_daily_files)
```
