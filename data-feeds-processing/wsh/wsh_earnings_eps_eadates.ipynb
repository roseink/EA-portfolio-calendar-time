```python
"""
filename: wsh_earnings_eps_eadates.ipynb

This is potentially draft code and prepare for what's
to come with getting this data from WSH.

"""
# Imports here
import pandas as pd
import glob, os
import numpy as np
import datetime
from datetime import timedelta
import time 
import re

# OneDrive path here
pth = "C:\\Users\clj585\OneDrive - Northwestern University\data_feeds\WSH\Data"
os.chdir(pth)
```

```python
"""
Get all of the earnings and EPS files in parquet format from directory. 

These files are the more recent history. Some overlaps because these 
daily files are throughout the 2021 year. 

The historic data stopped around 8/3/2021 - need de-duplicate later on. 

"""
# Use glob to pick out files by ending 
earn_files = glob.glob("*_ED.parquet")
eps_files = glob.glob("*_EPS.parquet")

# Save and concatenate all dataframes here by type 
earn_list = []
eps_list = []

# Loop over earnings parquet daily files
for file in earn_files:
    temp = pd.read_parquet(os.path.join(pth, file))
    # Very important step: create the "date" column to get it in snapshot form 
    temp['asofdate'] = re.findall(r'\d{1,2}\-\d{1,2}\-\d{4}', file)[0]
    earn_list.append(temp)
    
# Repeat loop over EPS parquet daily files 
for file in eps_files:
    temp = pd.read_parquet(os.path.join(pth, file))
    eps_list.append(temp)
    
# Join everything together in the lists
wsh_eps = pd.concat(eps_list)
wsh_earn = pd.concat(earn_list)
wsh_earn = wsh_earn[['event_id', 'company_id', 'stock_symbol', 'asofdate',
                    'isin','company_name', 'earnings_date', 'quarter', 
                    'fiscal_year', 'earnings_date_status', 'time_of_day', 
                    'audit_source', 'announce_datetime','prelim_earnings_date',
                     'updated', 'created']]
wsh_earn
```

```python
"""
Gathering files from history. 

History WSH files are saved in a folder in OneDrive, containing CSVs. 
Legacy hisotirc files were in Dropbox, under analogous names. 

The directory path in both places points to WSH/Data/hist_20210902. 

"""
histpth = pth + "/hist_20210902"
os.chdir(histpth)
earn_list = []
eps_list = []

earn_f = glob.glob("earnings*.csv")
eps_f = glob.glob("eps*.csv")

for f in earn_f:
    earn_list.append(pd.read_csv(os.path.join(histpth, f)))
    
for f in eps_f:
    eps_list.append(pd.read_csv(os.path.join(histpth, f)))
    
earnhist = pd.concat(earn_list)
epshist = pd.concat(eps_list)
earnhist
```

```python
"""
Rejoin historical data from above with previously concatenated parquet files 
to get data from 2006 to current. 

"""
wsh_earn = pd.concat([wsh_earn, earnhist])
wsh_eps = pd.concat([wsh_eps, epshist])

"""
Check and dedupe both data frames! 
This is due to overlapping dates hopefully. 

"""
wsh_earn = wsh_earn.drop_duplicates()

# Repeat same duplicate drop on wsh_eps (probably overlapping dates)
wsh_eps = wsh_eps.drop_duplicates()

wsh_earn
```

```python
"""
WSH earnings files - fixing column data types 
and subsetting columns taken care of here. 

"""
# Filter out columns and renaming
colskeep = ['event_id', 'company_id', 'stock_symbol', 
            'asofdate', 'isin', 'quarter', 'fiscal_year', 
            'earnings_date_status', 'time_of_day', 'announce_datetime', 
            'updated']
colmap = {"company_id" : "wsh_id", "stock_symbol":"hticker", "isin":"isin",
         "quarter": "fqtr", "fiscal_year" :"fyear", "announce_datetime":"exp_ea_datetime", 
         "asofdate":"date"}
wsh_earn2 = wsh_earn[colskeep]
wsh_earn2 = wsh_earn2.rename(columns = colmap)

# Define as strings
wsh_earn2[['wsh_id', "hticker", "fqtr", "exp_ea_datetime", "isin",
         "time_of_day", "date"]] = wsh_earn2[['wsh_id', "hticker", "fqtr", "exp_ea_datetime", "isin",
                                              "time_of_day", "date"]].astype("string")

# Input this as pandas date time stamp
wsh_earn2['exp_ea_datetime'] = pd.to_datetime(wsh_earn2['exp_ea_datetime'])
wsh_earn2
```

```python
"""
Earnings columns - more processing for columns that 
contain information about fiscal period. 

The quarter and fiscal year are available. Note that the foreign 
companies have their own reporting frequencies and notation, such as 
semiannual or annual basis (H, H1 and H2).

"""
wsh_earn2['fyear'] = wsh_earn2['fyear'].astype(float)
wsh_earn2['fyear'] = wsh_earn2['fyear'].astype("Int64")

# Adjust the NA stuff in the time of day column 
wsh_earn2['time_of_day'].fillna("Not Available", inplace=True)

# Adjust the time of day flags to be consistent 
wsh_earn2['time_of_day'] = np.where(wsh_earn2['time_of_day'] == "After Market", "PC", 
                                   wsh_earn2['time_of_day'])
wsh_earn2['time_of_day'] = np.where(wsh_earn2['time_of_day'] == "Before Market", "PO", 
                                   wsh_earn2['time_of_day'])
wsh_earn2['time_of_day'] = np.where(wsh_earn2['time_of_day'] == "During Market", "RH", 
                                   wsh_earn2['time_of_day'])
wsh_earn2['time_of_day'] = np.where(wsh_earn2['time_of_day'].str.contains("Unspecified", case=False), 
                                   "Not Available", wsh_earn2['time_of_day'])

wsh_earn2
```

```python
"""
Split apart the date & timestamp column (announce_datetime)
for the earnings dataset. Create exp_ea_date and exp_ea_time columns from this. 

"""
wsh_earn2['exp_ea_time'] = wsh_earn2['exp_ea_datetime'].dt.time
wsh_earn2['exp_ea_date'] = wsh_earn2['exp_ea_datetime'].dt.date

# Can absolutely handle the different formats of the date column (formerly asofdate)
wsh_earn2['date'] = pd.to_datetime(wsh_earn2['date']) 

# Generate hcusip and hcusip9 columns 
wsh_earn2['hcusip'] = wsh_earn2['isin'].str[2:10]
wsh_earn2['hcusip9'] = wsh_earn2['isin'].str[2:11]

wsh_earn2
```

```python
"""
Earnings data - WSH uses the dotted style to denote tickers share classes. 
Meanwhile, preferred shares and other international entities are denoted
as a prefix before the colon. 

Process and split tickers here.  

"""
# Longer itertive process here
wsh_earn2["step1"] = wsh_earn2['hticker'].str.lstrip("!")

# Then get rid of the suffix in step1 column to create step2
wsh_earn2[['step2', "suffix"]] = wsh_earn2['step1'].str.split("[/.]", n=1, expand=True)

# Get rid of the prefix before ":", but need to adjust
wsh_earn2[['prefix_unadj', "root_unadj"]] = wsh_earn2['step2'].str.split("[/:]", n=1, expand=True)

# If there is no colon, then the root should be equal to what's in the prefix_unadj column
wsh_earn2['root'] = np.where(wsh_earn2['hticker'].str.contains(":"),
                            wsh_earn2['root_unadj'],  wsh_earn2['prefix_unadj'])

# If there is no colon, then the prefix should be blank as well
wsh_earn2['prefix'] = np.where(wsh_earn2['hticker'].str.contains(":"), 
                               wsh_earn2['prefix_unadj'], "")
wsh_earn2['suffix'] = wsh_earn2['suffix'].fillna("")

wsh_earn2 = wsh_earn2[['date', 'event_id', 'wsh_id', 'hticker', 'fqtr', 'fyear', 
                      'time_of_day', 'exp_ea_datetime', 'exp_ea_date', 'exp_ea_time', 
                      'hcusip', 'hcusip9', 'suffix', 'root', 'prefix']]
```

```python
"""
WSH EPS files - fix column types and subsetting here. 

"""
# Saved desired columns and column name map 
colskeep = ['company_id', "stock_symbol", "announce_datetime", "fiscal_year", "isin",
           "quarter"]
colmap = {"company_id": "wsh_id", "stock_symbol":"hticker", "fiscal_year":"fyear", 
         "quarter":"fqtr", "announce_datetime" : "ea_datetime"}

# Subset and rename on EPS
wsh_eps = wsh_eps[colskeep]
wsh_eps = wsh_eps.rename(columns = colmap)

# String column types
wsh_eps[['wsh_id', "hticker", "fqtr", "isin",
         "ea_datetime"]]= wsh_eps[['wsh_id',"hticker", "fqtr","isin",
                                         "ea_datetime"]].astype('string')
wsh_eps
```

```python
"""
Date and time formatting - split the announcement date and timestamp combined
column into 2 different columns on EPS dataset. 

"""
# Fiscal year as float first and then to Int 64 (because of the NAN values handling)
wsh_eps['fyear'] = wsh_eps['fyear'].astype(float)
wsh_eps['fyear'] = wsh_eps['fyear'].astype("Int64")

# Date and time conversions involving actual EA announcement dates
wsh_eps['ea_datetime'] = pd.to_datetime(wsh_eps['ea_datetime'])
wsh_eps['ea_time'] = wsh_eps['ea_datetime'].dt.time
wsh_eps['ea_date'] = wsh_eps['ea_datetime'].dt.date
wsh_eps
```

```python
"""
Dotted ticker processing on EPS data frame, 
as well as taking care of prefixes. 

"""
# Longer itertive process here
wsh_eps["step1"] = wsh_eps['hticker'].str.lstrip("!")

# The check to make sure that this first step worked 
wsh_eps[wsh_eps['hticker'].str.contains("!")]
wsh_eps

# Then get rid of the colon ":" in step1 column to create step2
wsh_eps[['step2', "suffix"]] = wsh_eps['step1'].str.split("[/.]", n=1, 
                                                              expand=True)

# Last step!!! GRRR
wsh_eps[['prefix_unadj', "root_unadj"]] = wsh_eps['step2'].str.split("[/:]", n=1, 
                                                                     expand=True)

# Need to adjust things 
# If there is no colon, then the root should be equal to what's in the prefix_unadj column
wsh_eps['root'] = np.where(wsh_eps['hticker'].str.contains(":"),
                            wsh_eps['root_unadj'],  wsh_eps['prefix_unadj'])

# If there is no colon, then the prefix should be blank as well
wsh_eps['prefix'] = np.where(wsh_eps['hticker'].str.contains(":"), 
                               wsh_eps['prefix_unadj'], "")

# Fill out NA cells
wsh_eps['suffix'] = wsh_eps['suffix'].fillna("")
wsh_eps['prefix'] = wsh_eps['prefix'].fillna('')

# Handle isin to CUSIP conversions here
wsh_eps['hcusip9'] = wsh_eps['isin'].str[2:11]
wsh_eps['hcusip'] = wsh_eps['isin'].str[2:10]

wsh_eps = wsh_eps[["wsh_id","hticker","ea_datetime","fqtr","fyear",
                   "hcusip","hcusip9",
                   "ea_time","ea_date","prefix","root","suffix"]]

wsh_eps
```

```python
"""
Save both as parquet files.

"""
# Paths here 
pth1 = pth + "/wsh_earn_eadates.parquet"
pth2 = pth + "/wsh_eps_eadates.parquet"

# Save out as parquets in OneDrive
wsh_earn2.to_parquet(pth1)
wsh_eps.to_parquet(pth2)
```
